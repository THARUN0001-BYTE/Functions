**1. What is the difference between a function and a method in Python?**

**Answer: ** In Python, functions and methods are both used to perform specific tasks, but they differ in how they are defined and used.

**Function:**
A function is a block of code that performs a specific task.

It is not associated with any object or class.

Functions are defined using the def keyword.

Functions can be called independently.

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result)
```

**Method:**

A method is a function that is associated with an object or class.

It is defined inside a class and operates on instances (objects) of that class.

The first parameter of an instance method is usually self, which refers to the object calling the method.

```python
class Calculator:
    def add(self, a, b):
        return a - b

calc = Calculator()
result = calc.add(3, 5)
print(result)
```

**6. Explain the concept of generators in Python and how they are defined?**

Answer: Generators are special iterators that allow you to generate values on the fly instead of storing them in memory all at once. They are useful for working with large datasets, infinite sequences, or lazy evaluation.



```python
# yield inside a function:

def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
counter = count_up_to(3)

print(next(counter))
print(next(counter))
print(next(counter))
```

```python
#  Example: Generator for Even Numbers.

def even_numbers(limit):
    for num in range(limit):
        if num % 2 == 0:
            yield num

for e in even_numbers(6):
    print(e)
```

```python
# Generator Expressions:

gen = (x*x for x in range(5))
print(next(gen))
```

**9. Explain the purpose and usage of the map() function in Python?**

Answer:  The map() function is a built-in Python function used to apply a given function to every item in an iterable (like a list, tuple, etc.) and return a new map object (an iterator).

**10. What is the difference between map(), reduce(), and filter() functions in Python?**

Answer: These are functional programming tools in Python, used for transforming and processing iterables. Here's a clear comparison:

1.map(function, iterable)

.Applies a function to every item in an iterable.

```python
# Example:

numbers = [1, 25, 333, 444]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)
```

2.filter(function, iterable).

Filters items in an iterable based on a condition (function that returns True or False).

```python
# Example:

numbers = [1, 2, 3, 4, 5]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)

```

3.reduce(function, iterable)

Applies a function cumulatively to the items of an iterable, reducing it to a single value.

```python
# Example:

from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

**1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in  the list? **

Answer:

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
```

```python
# Example Usage:

my_list = [10, 15, 20, 25, 30]
result = sum_of_even_numbers(my_list)
print("Sum of even numbers:", result)
```

```python
# Optional (Using filter() and sum() with lambda):

def sum_of_even_numbers(numbers):
    return sum(filter(lambda x: x % 2 == 0, numbers))
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200?**

Answer:

```python
# Prime Check Function (1 to 200):

def is_prime(n):
    if n < 2 or n > 200:
        return False  # Not prime or out of range
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
# Example Usage:

print(is_prime(7))
print(is_prime(100))
print(is_prime(199))
print(is_prime(205))
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?**

Answer:

```python
# Fibonacci Iterator Class.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b
```

```python
fib = FibonacciIterator(10)

for num in fib:
    print(num, end=' ')
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent?**

Answer:

```python
# Generator Function: Powers of 2.

def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
```

**7. Implement a generator function that reads a file line by line and yields each line as a string?**

Answer:

```python
# Generator Function to Read a File Line by Line.

def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')
```

**10. Create a Python program that uses filter() to remove all the vowels from a given string?**

Answer:

```python
# Python Program Using filter():

def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

text = "Hello, World!"
result = remove_vowels(text)
print("String without vowels:", result)
```
